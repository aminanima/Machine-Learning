```python
import numpy as np
import scipy.linalg
from matplotlib import pylab as plt


def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def create_array(params):
    result = []
    if not isinstance(params, list):
        return result

    count = len(params)
    for i in range(count):
        result.append([params[i]**n for n in range(count)])
    return np.array(result)
```

```python
def create_vector(params):
    result = []
    if not isinstance(params, list):
        return result

    count = len(params)
    for i in range(count):
        result.append(f(params[i]))
    return result
```

```python
f_x = [f(x) for x in range(16)]
plt.plot(range(16), f_x)
plt.show()
```

```python
points = [1, 15]
A = create_array(points)
b = create_vector(points)
answer = scipy.linalg.solve(A, b)
print(answer)
f_x = answer
plt.plot(points, f_x)
plt.show()
```
