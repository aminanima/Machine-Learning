## Task1. сравнение предложений

```python
import re
import numpy as np
from scipy.spatial.distance import cosine
```

```python
#opening file
file_obj = open('sentences.txt')
```

```python
sentences = [] #empy list
for line in file_obj:
    line = line.lower()
    sentences.append(line.strip()) #add every line into list without spaces
print(sentences)
```

```python
#make token
list_of_words = []
list_of_all_words =[]
for i in range(len(sentences)):
    words = re.split('[^a-z]', sentences[i])
    words[:] = (word for word in words if len(word) != 0) #remove empty words
    list_of_words.append(words)
    list_of_all_words = list(set(list_of_all_words + words)) #remove repetitions and make one list
print(list_of_words)
print(list_of_all_words)
file_obj.close()
```

```python
dict_words = {x: y for x, y in zip(range(len(list_of_all_words)), list_of_all_words)}
print(dict_words)
```

```python
number_of_occurs = [] #matrix
for i in range(len(list_of_words)):
    number = [] # show how many times occurs in sentence
    for key in dict_words.keys():
        num = list_of_words[i].count(dict_words[key])
        number.append(num)
    number_of_occurs.append(number)
print(number_of_occurs)
```

```python
matrix = np.array(number_of_occurs)
print(matrix)
```

```python
d = {}
for i in range (22):
    d[i] = cosine(matrix[0], matrix[i])
print(d)
```

```python
d_copy = d.copy()
del d_copy[0]
min1 = min(d_copy.values())
del d_copy[6]
min2 = min(d_copy.values())
print(min1, min2)
```

```python
def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
first = get_key(d, min1)
second = get_key(d, min2)
print(first)
print(second)
```

## Task 2. аппроксимация функции

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 15, .5)
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

plt.plot(x, y)
plt.show()
```

```python
x1_1 = 1
x1_2 = 15

y1_1 = np.sin(x1_1 / 5) * np.exp(x1_1 / 10) + 5 * np.exp(-x1_1 / 2)
y1_2 = np.sin(x1_2 / 5) * np.exp(x1_2 / 10) + 5 * np.exp(-x1_2 / 2)

a1 = np.array([[1, x1_1], [1, x1_2]])
b1 = np.array([y1_1, y1_2])
w1 = np.linalg.solve(a1, b1)
print(w1)
```

```python
x = np.arange(1, 15, .5)
y_1 = sum(w1[i] * x**i for i in range(2))
plt.plot(x, y, '-', x, y_1, '-')
plt.show()
```

```python
x2_1 = 1
x2_2 = 8
x2_3 = 15

y2_1 = np.sin(x2_1 / 5) * np.exp(x2_1 / 10) + 5 * np.exp(-x2_1 / 2)
y2_2 = np.sin(x2_2 / 5) * np.exp(x2_2 / 10) + 5 * np.exp(-x2_2 / 2)
y2_3 = np.sin(x2_3 / 5) * np.exp(x2_3 / 10) + 5 * np.exp(-x2_3 / 2)

a2 = np.array([[1, x2_1, x2_1**2], [1, x2_2, x2_2**2], [1, x2_3, x2_3**2]])
b2 = np.array([y2_1, y2_2, y2_3])
w2 = np.linalg.solve(a2, b2)
```

```python
x = np.arange(1, 15, .5)
y_1 = sum(w1[i] * x**i for i in range(2))
y_2 = sum(w2[i] * x**i for i in range(3))
plt.plot(x, y, '-', x, y_1, '-', x, y_2, '-')
plt.show()
```

```python
x3_1= 1
x3_2 = 4
x3_3 = 10
x3_4 = 15

y3_1 = np.sin(x3_1 / 5) * np.exp(x3_1 / 10) + 5 * np.exp(-x3_1 / 2)
y3_2 = np.sin(x3_2 / 5) * np.exp(x3_2 / 10) + 5 * np.exp(-x3_2 / 2)
y3_3 = np.sin(x3_3 / 5) * np.exp(x3_3 / 10) + 5 * np.exp(-x3_3 / 2)
y3_4 = np.sin(x3_4 / 5) * np.exp(x3_4 / 10) + 5 * np.exp(-x3_4 / 2)

a3 = np.array([[1, x3_1, x3_1**2, x3_1**3], [1, x3_2, x3_2**2, x3_2**3], [1, x3_3, x3_3**2, x3_3**3], [1, x3_4, x3_4**2, x3_4**3]])
b3 = np.array([y3_1, y3_2, y3_3, y3_4])
w3 = np.linalg.solve(a3, b3)
print(w3)
```

```python
x = np.arange(1, 15, .5)
y_1 = sum(w1[i] * x**i for i in range(2))
y_2 = sum(w2[i] * x**i for i in range(3))
y_3 = sum(w3[i] * x**i for i in range(4))
plt.plot(x, y, '-', x, y_1, '-', x, y_2, '-', x, y_3, '-')
plt.show()
```

```python
coef = list((str(round(w3[i], 2)) for i in range(4)))
answer = ', '.join(coef)
print(answer)
```

## Task 3. Минимизация гладкой функции

```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution
from math import exp, sin
```

```python
def func (x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
```

```python
x = []
y = []
for num in range (1, 31):
    f = func(num)
    x.append(num)
    y.append(f) 

plt.plot(x, y)
plt.show()
```

```python
print(minimize(func, 30, method='BFGS'))
```

```python
print(differential_evolution(func, [(1, 30)]))
```

```python
def h_func (x):
    return int(func(x))

t = []
z = []
for num in range (1, 31):
    h = int(func(num))
    t.append(num)
    z.append(h) 

plt.plot(t, z)
plt.show()
```

```python
print(minimize(h_func, 30, method='BFGS'))
```

```python
print(differential_evolution(h_func, [(1, 30)]))
```
